# Data Visualization (Visualisasi Data)

**Univariate Analysis**

Charges (Biaya)

```python
plt.figure(figsize=(10,6))
sns.distplot(df.charges,color='r')
plt.title('Distribution of Charges',size=18)
plt.xlabel('Charges',size=15)
plt.ylabel('Density',size=15)
plt.show()
```

Age (Usia)

```python
plt.figure(figsize=(10,6))
sns.histplot(df.age)
plt.title('Age Distribution',size=18)
plt.xlabel('Age',size=15)
plt.ylabel('Count',size=15)
plt.show()
```

BMI (Indeks Massa Tubuh)

```python
plt.figure(figsize=(10,6))
plt.hist(df.bmi,color='y')
plt.title('BMI Distribution',size=18)
plt.show()
```

Gender (Jenis Kelamin)

```python
plt.figure(figsize=(10,6))
sns.countplot(x='sex', data=df)
plt.title('Total Number of Male and Female',size=18)
plt.xlabel('Sex',size=15)
plt.show()
plt.show()
```

Children (Anak-anak)

```python
plt.figure(figsize=(10,6))
sns.countplot(x='children', data=df)
plt.title('Total Number of Children',size=18)
plt.xlabel('Children',size=15)
plt.ylabel('Count',size=15)
plt.show()
```

Smoker (Perokok)

```python
plt.figure(figsize=(10,6))
sns.countplot(x='smoker', data=df)
plt.title('Total Number of Smoker',size=18)
plt.xlabel('Smoker',size=15)
plt.ylabel('Count',size=15)
plt.show()
```

```python
df.smoker.value
```

Region (Wilayah)

```python
plt.figure(figsize=(10,6))
sns.countplot(x='region', data=df)
plt.title('Total Number of Region',size=18)
plt.xlabel('Region',size=15)
plt
```

**Bivariate Analysis**

Age vs Charges (Usia vs Biaya)

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='charges', color = 'r', data=df)
plt.title('Age vs Charges',size=18)
plt.xlabel('Age',size=15)
plt.ylabel('Charges',size=15)
plt.show()
```

Smoker vs Charges (Perokok vs Tuduhan)

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title('Smoker vs Charges',size=18)
```

```python
sns.pairplot(df,
             markers='+',
             diag_kind='kde',
             kind='reg',
             plot_kws={'line_kws':{'color':'#aec6cf'},
                       'scatter_kws':{'alpha':0.7,
                                      'color':'red'}},
             corner=True)
```

**Multivariate Analysis**

Correlation (Korelasi)

```python
plt.figure(figsize=(10,6))
numerical_df = df.select_dtypes(include=['number'])
sns.heatmap(numerical_df.corr(), annot=True, square=True)
plt.title('Correlation Between Variables', size=18)
plt.xticks(size=13)
plt.yticks(size=13)
plt.show()
```
