```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
#import warnings
#warnings.filterwarnings('ignore'
```

```python
df = df.drop(['Engine Fuel Type','Market Category', 'Vehicle Style','Popularity','Number of Doors'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
numerical_df = df.select_dtypes(include=np.number)

Q1 = numerical_df.quantile(0.25)
Q3 = numerical_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Reindex Q1, Q3, and IQR to match df's index before filtering
numerical_cols = df.select_dtypes(include=np.number).columns
Q1 = Q1.reindex(numerical_cols, fill_value=0)
Q3 = Q3.reindex(numerical_cols, fill_value=np.inf)
IQR = IQR.reindex(numerical_cols, fill_value=0)  # Reindex IQR as well


# Filter the DataFrame using aligned quantiles, only for numerical columns
filtered_df = df[~((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
filtered_df.shape
```
